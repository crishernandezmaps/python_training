```python
import pandas as pd
```

```python
# Reading data: CSV
csv_hosted_on_github = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv'
df_csv_from_Github = pd.read_csv(csv_hosted_on_github,sep=',')
df_CSV_from_Github.head()
```

```python
# Reading data: HTML Tables
html_table_hosted_on_github = 'https://en.wikipedia.org/wiki/List_of_English_words_containing_Q_not_followed_by_U'
df_table_from_html = pd.read_html(html_table_hosted_on_github)
first_table_element_on_array_of_tables = df_table_from_html[0]
first_table_element_on_array_of_tables.tail()
```

```python
# Exporting data frames as Excel
df_csv_from_Github.to_excel('table_from_csv.xlsx')
first_table_element_on_array_of_tables.to_excel('table_from_html.xlsx')
```

```python
# HTML
df_html_list = pd.read_html('https://en.wikipedia.org/wiki/List_of_mountains_by_elevation') # list of dataframes

df_html_wrong_header = df_html_list[0] # we need the first dataframe in this example

df_html_wrong_header.columns = df_html_wrong_header.iloc[0] # select first index content as indexes

df_html = df_html_wrong_header.reindex(df_html_wrong_header.index.drop(0)) # finally we drop Index 0 to avoid duplicates

df_html
```
